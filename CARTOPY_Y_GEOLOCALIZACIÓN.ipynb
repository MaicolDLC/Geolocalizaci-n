# Geolocalización Utilizando GeoPy 

---
GeoPy es una librería de Python que facilita la localización de las coordenadas geográficas de direcciones, ciudades, países y puntos de referencia en todo el mundo.

```python
cities = ['Buenos Aires', 'Río de Janeiro', 'Santiago de Chile', 'Bogota', 'Caracas','Lima Perú','La Paz Bolivia','Paita Perú']

fig , ax = plt.rcParams['figure.figsize'] = (10,10)
ax = plt.axes(projection=cartopy.crs.PlateCarree())
ax.stock_img()
ax.coastlines(resolution='10m',color='black', linewidth=1)
ax.add_feature(cfeature.LAND, zorder=0, edgecolor='black')
ax.add_feature(cfeature.BORDERS, edgecolor='black', linewidth=1)
#---------------------------------------------------------------------------------------------------------
ax.set_extent((-90, -30, -60, 15))
grids = ax.axes.gridlines(color='black', alpha=0.25, linestyle='--', linewidth = 0.5,
                              crs=ccrs.PlateCarree(),
                              draw_labels=False)

ax.set_xlabel('LONGITUD',fontdict={'fontsize':20,
                                  'color':'black'})
ax.set_ylabel('LATITUD',fontdict={'fontsize':20,
                                  'color':'black'})

ax.set_xticks([-90,-70,-50,-30])
ax.set_yticks([-60,-40,-20,0])
#---------------------------------------------------------------------------------------------------------
location = Nominatim(user_agent="location")
for city in cities:    
  address = location.geocode(city)
  x, y = np.meshgrid(address.longitude, address.latitude) 
  ax.plot(x,y,color='red',marker='*',markersize='15')
```
